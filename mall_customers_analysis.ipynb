```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv(r"E:\github_files\Mall_Customers.csv")
```

```python
customers.head()
```

```python
customers.describe() #to see descriptive stats of each variable
```

```python
#exploring the data
sns.countplot(x='Gender', data=customers)
plt.title('Distribution Of Gender')
```

```python
#creating histogram of age
plt.hist('Age',data=customers,bins=35);
plt.title('Distribution of Age');
plt.xlable('Age');
```

```python
plt.hist('Age', data=customers[customers['Gender'] == 'Male'],alpha=0.5, label='Male')
plt.hist('Age', data=customers[customers['Gender'] == 'Female'],alpha=0.5, label='Female')
plt.title('Distribution of Age by Gender')
plt.legend()
```

```python
#same above both plots with anuual income variable
```

```python
male_customers = customers[customers['Gender'] == 'Male']
female_customers = customers[customers['Gender'] == 'Female']

#print average spending score for men and women
print(male_customers['Spending Score (1-100)'].mean())
print(female_customers['Spending Score (1-100)'].mean()) #women earned less but spend more at this mall
```

```python
#scatter plot
sns.scatterplot('Age', 'Annual Income (k$)',hue='Gender', data=customers)
plt.title('Age to income, colored by gender')
```

```python
sns.heatmap(customers.corr(),annot=True)
```

```python
sns.scatterplot('Age', 'Spending Score (1-100)',hue='Gender', data=customers)
```

```python
sns.heatmap(female_customers.corr(), annot=True)
plt.title('Correlation Heatmap - Female')
```

```python
sns.heatmap(male_customers.corr(), annot=True)
plt.title('Correlation Heatmap-Male')
```

```python
#when find out with heatmap age more strongly affects spending score for women in this case
#now we can zoom in on the women spending score to age relationship with the lmplot
sns.lmplot('Age', 'Spending Score (1-100)', data=female_customers)
plt.title('Age to spending score, female only')
```

```python
plt.hist('Age',data=female_customers)
plt.title('Age distribution and spending score for women')
plt.legend() #shows that women at the age around 25 to 30 spends more money 
```

```python
sns.scatterplot('Annual Income (k$)', 'Spending Score (1-100)',hue='Gender',data=customers)
plt.title('Annual income to spending score,colored by Gender')
```

```python
sns.lmplot('Annual Income (k$)', 'Spending Score (1-100)',data=customers)
plt.title("plot b|w annual income and the spending score")
```
